```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Download data



```python
games = pd.read_csv('C:\\Users\\MacBookPro\\Desktop\\games.csv')  
developers = pd.read_csv('C:\\Users\\MacBookPro\\Desktop\\developers.csv')  
platforms = pd.read_csv('C:\\Users\\MacBookPro\\Desktop\\platforms.csv')  
genres = pd.read_csv('C:\\Users\\MacBookPro\\Desktop\\genres.csv')  
scores = pd.read_csv('C:\\Users\\MacBookPro\\Desktop\\scores.csv')
```

# User ratings and data overview

```python
print("Games Data:")
print(games.head())
print("\nDevelopers Data:")
print(developers.head())
print("\nPlatforms Data:")
print(platforms.head())
print("\nGenres Data:")
print(genres.head())
print("\nScores Data:")
print(scores.head())
```

# Data cleaning
# Check if there are missing values

```python
print("Missing values in games data:")
print(games.isnull().sum())
```

```python
print("Missing values in games data:")
print(games.isnull().sum())
```

# If there are missing values, you can choose to delete or fill them
# Here is a simple example, delete rows containing missing values

```python
games = games.dropna()
```

# Basic statistical analysis

```python
print("Average game rating:")
print(games['rating'].mean())

print("Top 10 most played games:")
print(games.nlargest(10, 'plays')[['name', 'plays']])
```

# data visualization
# Game rating distribution

```python
sns.histplot(games['rating'], kde=True)
plt.title("Distribution of Game Ratings")
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.show()
```

# Game type analysis

```python
genre_counts = genres['genre'].value_counts()
sns.barplot(x=genre_counts.values, y=genre_counts.index)
plt.title("Number of Games by Genre")
plt.xlabel("Number of Games")
plt.ylabel("Genre")
plt.show()
```

# Platform analysis

```python
platform_counts = platforms['platform'].value_counts()
sns.barplot(x=platform_counts.values, y=platform_counts.index)
plt.title("Number of Games by Platform")
plt.xlabel("Number of Games")
plt.ylabel("Platform")
plt.show()
```
